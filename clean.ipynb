```python
data["FL_DATE"] = pd.to_datetime(data["FL_DATE"], format="%m/%d/%Y %I:%M:%S %p", errors='coerce')
```

```python
key_cols = [
    "FL_DATE",
    "OP_UNIQUE_CARRIER",
    "OP_CARRIER_FL_NUM",
    "ORIGIN_AIRPORT_ID",
    "DEST_AIRPORT_ID"
]
```

```python
data[key_cols].isnull().sum()
```

```python
data = data.dropna(subset = key_cols)
```

```python
# Duplicates
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
dup = data.duplicated(subset = key_cols, keep=False)
print("Total rows:", len(data))
print("Rows with duplicated keys:", dup.sum())
```

```python
data = data.drop_duplicates(subset = key_cols, keep = "first")
```

```python
print("Cancelled values:", data["CANCELLED"].unique())
print("Diverted values:", data["DIVERTED"].unique())
```

```python
data.loc[:, "CANCELLED"] = pd.to_numeric(data["CANCELLED"], errors = 'coerce')
data.loc[:, "DIVERTED"] = pd.to_numeric(data["DIVERTED"], errors = 'coerce')
data.loc[:, "CANCELLED"] = data["CANCELLED"].fillna(0).astype(int)
data.loc[:, "DIVERTED"] = data["DIVERTED"].fillna(0).astype(int)
```

```python
string_cols = [
    "OP_UNIQUE_CARRIER",
    "ORIGIN_STATE_ABR",
    "DEST_STATE_ABR",
    "ORIGIN_CITY_NAME",
    "DEST_CITY_NAME",
    "CANCELLATION_CODE",
    "DIV1_AIRPORT",
    "DIV2_AIRPORT"
]

for col in string_cols:
    if col in data.columns:
        data.loc[:, col] = data[col].astype(str).str.strip()
        data.loc[:, col] = data[col].replace(["nan","NAN"], pd.NA)
```

```python
code_cols = [
    "OP_UNIQUE_CARRIER",
    "ORIGIN_STATE_ABR",
    "DEST_STATE_ABR",
    "CANCELLATION_CODE",
    "DIV1_AIRPORT",
    "DIV2_AIRPORT"
]
for col in code_cols:
    if col in data.columns:
        data.loc[:, col] = data[col].str.upper()
```

```python
city_cols = [
    "ORIGIN_CITY_NAME",
    "DEST_CITY_NAME",
]
for col in city_cols:
    if col in data.columns:
        data.loc[:, col] = data[col].str.title()
```

```python
for col in code_cols:
    if col in data.columns:
        print(data[col].unique())
```

```python
# Validation
assert data.duplicated(subset = key_cols).sum() == 0, "Duplicates found in key columns"
assert data[key_cols].isnull().sum().sum() == 0, "Nulls found in key columns"
assert set(data["CANCELLED"].unique()).issubset({0,1}), "Cancelled values are not 0 or 1"
assert set(data["DIVERTED"].unique()).issubset({0,1}), "Diverted values are not 0 and 1"
assert str(data["FL_DATE"].dtype).startswith("datetime"), "Flight Date is not a datetime"
for col in string_cols:
    if col in data.columns:
        assert "NaN" not in data[col].astype(str).unique(), f"{col} contains NaN"
```
